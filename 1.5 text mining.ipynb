```python
# Sentence tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10, cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

filtered_words = []  # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
from nltk import FreqDist  # make sure this is included

# Create a new FreqDist for filtered_words
Dist_words_filter = FreqDist(filtered_words)
print(Dist_words_filter)
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
Dist_words_filter.plot(10, cumulative=False)  
plt.show()
```

```python
# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization
Tokenized_word_2 = word_tokenize(sans_punc)
print(Tokenized_word_2) 
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(Tokenized_word_2) 
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10, cumulative=False)  
plt.show()
```

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The']

filtered = []
for word in Tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top20=df_t.nlargest(20, 'Words')
```

## Creating Bar Plots

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top20).set_title("Alice's Adventures in Wonderland - top 20 word types used")
```

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences') 
```

```python
plt.figure(figsize=(10, 5))

# Set a better color palette
sns.set_palette("Blues_d")  # A darker shade of blue

# Create the bar plot
sns.barplot(x="Word", y="Occurrences", saturation=0.9, data=top10).set_title("Alice's Adventures in Wonderland - Most Frequently Used Nouns")

plt.show()
```

## Bar Plots for Text Mining Specific Information

```python
listToStr = ' '.join([str(elem) for elem in filtered])

print(listToStr)
```

```python
# import the characters data set as a pandas dataframe
chars=pd.read_csv('/Users/yasersouri/Desktop/data analysis/specialization 1/Alice_Network_Analysis/character_df.csv',index_col=0)
```

```python
# Create a count for the main characters

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
chars['character_alias'] = chars['character'].apply(lambda x: x.rsplit(' ',1)[-1])
```

```python
chars
```

```python
chars_list = chars['character_alias'].to_list()
```

```python
dict_of_counts = {d : all_counts[d] for d in chars_list}
```

```python
#Search for the names from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}
df = pd.DataFrame(dct)
```

```python
df
```

```python

# Reshape the dataframe
df = df.melt(var_name='CHARACTER', value_name='TIMES MENTIONED')

# Show the final table
print(df)
```

```python

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['CHARACTER'], df['TIMES MENTIONED'], color='skyblue')

# Labels and title
plt.title('Character Mentions')
plt.xlabel('CHARACTER')
plt.ylabel('TIMES MENTIONED')

# Show the plot
plt.tight_layout()
plt.show()
```
